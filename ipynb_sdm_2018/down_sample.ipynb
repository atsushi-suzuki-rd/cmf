```python
import sys,os
sys.path.append(os.pardir)
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
dir_path = 'D:/UCI/gas'
original_file_name = 'ethylene_methane.txt'
original_file_path = dir_path + '/' + original_file_name
```

```python
data = pd.read_csv(original_file_path, delim_whitespace=True)
```

```python
output_file_name = 'ethylene_methane_down_sampled.txt'
output_file_path = dir_path + '/' + output_file_name
```

```python
max_time = int(data.max()['Time_(seconds)'])
max_time
```

```python
down_sampled_idxs = [data[(data['Time_(seconds)']>t-1) & (data['Time_(seconds)']<=t)].index[-1] for t in range(max_time)]
```

```python
down_sampled_data = data.ix[down_sampled_idxs]
```

```python
down_sampled_data
```

```python
down_sampled_data.to_csv(output_file_path)
```
